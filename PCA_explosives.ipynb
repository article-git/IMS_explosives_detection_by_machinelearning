# Data pre-processing

### autoscaling

```python
# autoscale the data
from sklearn.preprocessing import scale
X_autosc = scale(X.values)
X_autosc
```

### SNV

```python
# Write the SNV function (is, actually, like autoscaling by row)
def snv(input_data):
  
    # Define a new array and populate it with the corrected data  
    output_data = np.zeros_like(input_data)
    for i in range(input_data.shape[0]):
 
        # Apply correction
        output_data[i,:] = (input_data[i,:] - np.mean(input_data[i,:])) / np.std(input_data[i,:])
 
    return output_data
```

```python
# Compute the SNV on np data
Xsnv = snv(X.values)
Xsnv
```

### smoothing

```python
# Preprocessing with Savitzki-Golay - smoothing, defining the window, the order and the use of derivatives
from scipy.signal import savgol_filter
X_savgol = savgol_filter(X, 15, polyorder = 6, deriv=0)
X_savgol
```

### smoothing + snv  

```python
# We can also combine the preprocessing strategies together: Savitzki-Golay - smoothing + SNV
X_savgol = savgol_filter(X, 15, polyorder = 6, deriv=0)
X_snv_savgol = snv(X_savgol)
X_snv_savgol
```

### smoothing + autoscaling

```python
X_savgol = savgol_filter(X, 15, polyorder = 6, deriv=0)
X_autosc_savgol = scale(X_savgol)
X_autosc_savgol
```

### snv + autoscaling

```python
X_snv_autosc = scale(Xsnv)
X_snv_autosc
```

### smoothing + snv + autoscaling

```python
X_snv_savgol_autosc = scale(X_snv_savgol)
X_snv_savgol_autosc
```

```python
# Get PCA scores
T = scores.iloc[:,0:4]
# Get PCA loadings
P = loadings.iloc[:,0:4]
# Calculate error array
Err = X - np.dot(T,P.T)
# Calculate Q-residuals (sum over the rows of the error array)
Q = np.sum(Err**2, axis=1)
# Calculate Hotelling's T-squared (note that data are normalised by default)
Tsq = np.sum((T/np.std(T, axis=0))**2, axis=1)
```

```python
# set the confidence level
conf = 0.95
ncomp = 4
```

```python
import scipy.stats
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
Tsq_conf = mean_confidence_interval(Tsq.values, confidence=0.95)
Tsq_conf = Tsq_conf[2]
Q_conf = mean_confidence_interval(Q.values, confidence=0.95)
Q_conf = Q_conf[2]
```

```python
# Create a dataframe using only T2 and Q-residuals
hot_q_data = {'T2': Tsq, 'Qres': Q, 'Class': scores.Class}  
hot_q_data = pd.DataFrame(hot_q_data, index = scores.index)
hot_q_data
```

```python
# Plot the Hotelling T2 vs Q-residuals plot
fig = px.scatter(hot_q_data, x="T2", y="Qres", hover_data={'Sample': (hot_q_data.index)},  color = "Class")
fig.add_hline(y=abs(Q_conf),line_dash="dot", line_color='Red')
fig.add_vline(x=Tsq_conf,line_dash="dot", line_color='Red')
fig.update_traces(textposition='top center')
fig.update_layout(
    height=600,
    width=800,
    title_text="Hotelling's T2 vs Q-residuals")
fig.show()
```
